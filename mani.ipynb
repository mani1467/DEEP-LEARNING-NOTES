### ANN (ARTIFICIAL NEURAL NETWORK):
* AN Artificial neural network is an computational model(means mathematical and algorithmic strucutre)  inspired by the way biological neural network in the human brain work it is a key technology in deep learning and mechine learning.
* ANN are designed to recognize patterns and relationships in data through a process of learning,generalizing from examples.

### How does ANN Work.
* An ANN consists of layer of interconnceted nodes(neurons),each connection has weight and each neuron has an associated bias.the network process information through these connection and neurons using mathematical functions.

### Working of ANN
* Forward propagation
* loss calculation
* Back propagation
* weight update(optimization)
* Iteration

#### Clean& PReprocess Data

```python
df.drop('customerID',axis=1,inplace=True)  # drop customerid it is not useful
```

```python
#  convert totalcharges to numeric(has blanks)
df['TotalCharges']=pd.to_numeric(df['TotalCharges'],errors='coerce')
df['TotalCharges'].fillna(df['TotalCharges'].median(),inplace=True)

#convert target to binary
df['Churn']=df['Churn'].map({'yes':1,'No':0})
```

#### Encode Categorical VAriables

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# identify categorical columns
cat_cols=df.select_dtypes(include=['object']).columns
```

```python
#label encode binary columns
binary_cols=[col for col in cat_cols if df[col].nunique()==2]
for col in binary_cols:
  df[col]=LabelEncoder().fit_transform(df[col])
```

```python
#one hot encode others
df=pd.get_dummies(df,columns=[col for col in cat_cols if col not in binary_cols])
```

```python
df.dtypes
```

## Feature Scaling and Splitting


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
x=df.drop('Churn',axis=1).values
y=df['Churn'].values
```

```python
scaler=StandardScaler()
x=scaler.fit_transform(x)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

## Build the ANN model

```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader,TensorDataset
```

```python
# convert to tensors
x_train_t=torch.tensor(x_train,dtype=torch.float32)
y_train_t=torch.tensor(y_train,dtype=torch.float32).view(-1,1)

x_test_t=torch.tensor(x_test,dtype=torch.float32)
y_test_t=torch.tensor(y_test,dtype=torch.float32).view(-1,1)

train_data=TensorDataset(x_train_t,y_train_t)
train_loader=DataLoader(train_data,batch_size=64,shuffle=True)
```

## Define the model

```python
class ANNModel(nn.Module):
  def __init__(self,input_dim):
    super().__init__()
    # self.fc1=nn.Linear(input_dim,64)
    # self.fc2=nn.Linear(64,32)
    # self.out=nn.Linear(32,1)
    # self.relu=nn.ReLU()
    # self.dropout=nn.Dropout(0.3)
    self.layers=nn.Sequential(
        nn.Linear(input_dim,128),
        nn.BatchNorm1d(128),
        nn.ReLU(),
        nn.Dropout(0.4),
        nn.Linear(128,64),
        nn.Dropout(0.3),
        nn.Linear(64,1)
    )

  def forward(self,x):
    # x=self.relu(self.fc1(x))
    # x=self.dropout(x)
    # x=self.relu(self.fc2(x))
    # x=self.out(x)
    x = self.layers(x)
    return x

model=ANNModel(x.shape[1])
```

### Train the Model

```python
optimizer = torch.optim.Adam(model.parameters(), lr=0.0005, weight_decay=1e-5)

loss_fn=nn.BCEWithLogitsLoss()

epochs=50
for epoch in range(epochs):
  model.train()
  total_loss=0
  for x_batch,y_batch in train_loader:
    optimizer.zero_grad()
    y_pred=model(x_batch)
    loss=loss_fn(y_pred,y_batch)
    loss.backward()
    optimizer.step()
    total_loss+=loss.item()
  print(f"epoch {epoch+1} | loss: {total_loss/len(train_loader):.4f}")


```

### Evaluate the model

```python
from sklearn.metrics import accuracy_score,classification_report
import warnings
warnings.filterwarnings('ignore')
```

```python
model.eval()
with torch.no_grad():
    # 1. Predict using model
    raw_preds = model(x_test_t)

    # 2. Replace any NaNs in predictions (safety step)
    raw_preds = torch.nan_to_num(raw_preds, nan=0.0)

    # 3. Apply sigmoid and round to get class labels (0 or 1)
    preds = torch.sigmoid(raw_preds).round()

    # 4. Convert to NumPy arrays
    preds_np = preds.numpy().astype(int)
    y_test_np = y_test_t.numpy().astype(int)

    # 5. Evaluate
    acc = accuracy_score(y_test_np, preds_np)
    print(f"Test Accuracy: {acc * 100:.2f}%")
    print(classification_report(y_test_np, preds_np, target_names=["No Churn", "Churn"]))
```
